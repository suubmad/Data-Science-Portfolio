# Create model

```python
y = df["Type"].values
x = df.drop("Type", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test =  train_test_split(x, y, test_size=0.3, random_state=15)
```

## scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

# first unit 30 because we have 30 columns
model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
# we add "sigmoid" = 0 & 1 for Classification
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=750, validation_data=(x_test, y_test), verbose=1)
```

```python
model.history.history
```

```python
modelLoss = pd.DataFrame(model.history.history)
```

```python
modelLoss.plot()
```

## earlystopping

- taking precautions for overfitting
- this process will stop when it gets to epoch we need

```python
model = Sequential()

# first unit 30 because we have 30 columns
model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
# we add "sigmoid" = 0 & 1 for Classification
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[earlyStopping])
```

```python
modelLoss2 = pd.DataFrame(model.history.history)
```

```python
modelLoss2.plot()
```

## dropout

- partial deletion of possible overfitting between layers

```python
model = Sequential()

# first unit 30 because we have 30 columns
model.add(Dense(units=30,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

# we add "sigmoid" = 0 & 1 for Classification
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[earlyStopping])
```

```python
modelLoss3 = pd.DataFrame(model.history.history)
```

```python
modelLoss3.plot()
```
